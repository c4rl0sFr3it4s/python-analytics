```python
# carregar os pacotes

from scipy import misc
import numpy as np
import matplotlib.cm as cm
import tensorflow as tf
import os, sys, inspect
from datetime import datetime
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from modulos import utils
from modulos.utils import testResult

from tensorflow.python.framework import ops
from sklearn.metrics.classification import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
import warnings
%matplotlib inline
```

```python
# parametros para fazer o reset dos grafos

warnings.filterwarnings("ignore")
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
ops.reset_default_graph()
```

```python
# definindo as emoções

emotion = {0:'anger', 
           1:'disgust',
           2:'fear',
           3:'happy',
           4:'sad',
           5:'surprise',
           6:'neutral'}
```

```python
# conversão de RGB para Gray-escala de cinza

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
# criando a sessão interativa

sess = tf.InteractiveSession()
```

```python
# restaurando o modelo

new_saver = tf.train.import_meta_graph('modelo/model.ckpt-900.meta')
new_saver.restore(sess, 'modelo/model.ckpt-900')
tf.get_default_graph().as_graph_def()
```

```python
# definindo o que será imagem de entrada e de saída

x = sess.graph.get_tensor_by_name("input:0")
y_conv = sess.graph.get_tensor_by_name("output:0")
```

```python
# carregar imagem

img = mpimg.imread('images_teste/image04.jpg')     
gray = rgb2gray(img)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
# redimensionando a imagem para utilizar o mesmo tamanho do treinamento

image_0 = np.resize(gray,(1,48,48,1))
tResult = testResult()
num_evaluations = 50
```

```python
# saída detectar qual emoção prevalece na foto

for i in range(0, num_evaluations):
    result = sess.run(y_conv, feed_dict={x:image_0})
    label = sess.run(tf.argmax(result, 1))
    label = label[0]
    label = int(label)
    tResult.evaluate(label)
tResult.display_result(num_evaluations)
```
