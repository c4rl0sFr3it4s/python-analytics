## Gráficos a partir do NumPy

```python
# importando pacote matpotlib, módulo pyplot 
# por convenção é apelidado de plt
# importando pacotes NumPy
# por convenção é apelidade np
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# scatter - dispensar
# definindo eixo x e y
plt.scatter(np.arange(50), np.random.randn(50))
# mostrando o gráfico
plt.show()
```

```python
# criando gráfico paralelo,  a forma de avaliar os dados com gráfico
# um do lado do outro
# criando sub_plot 
# modelando o plot
# start, stop, steps

fig = plt.figure()
# eixo x, gráfico 1 
ax1 = fig.add_subplot(1, 2, 1)
# eixo y
ax1.plot(np.random.randn(50), color = 'r')

# eixo x, gráfico 2
ax2 = fig.add_subplot(1, 2, 2)
# eixo y
ax2.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```python
# plots diversos
# subplots com 2 linhas, 3 colunas
# seguimentando a área scartternada

_, ax = plt.subplots(2, 3)
# gráfico de linha
ax[0, 1].plot(np.random.randn(50), color = 'green', linestyle = '-')
# histograma
ax[1, 0].hist(np.random.randn(50))
# scatter - dispernsa
ax[1, 2].scatter(np.arange(50), np.random.randn(50), color = 'r')
# mostrar 
plt.show()
```

```python
# controle dos eixos
fig, axes = plt.subplots(1, 3, figsize=(17, 4))
x1 = 0
x1 ** 2
x2 = 0
x2 ** 3
axes[0].plot(x1, x2)
axes[0].set_title('Eixos com range padrão.')

axes[1].plot(x1, x2)
axes[1].axis('tight')
axes[1].set_title('Eixos menores')

axes[2].plot(x1, x2)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title('Eixos customizados');
```

```python
# um outra coisa importante é a escala
# a escala pode afetar de maneira determinante
# a interpretação do gráfico
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x1, x2)
axes[0].set_title('Escala padrão.')

axes[1].plot(x1, x2)
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)');
```

É comum você encontrar na mídia gráficos totalmente dedurpados
o gráfico não representa realmente a informação porque não foi obtido de maneira de escala ou de maneira acidental ou de maneira proposital dependendo do portal de notícia.

```python
# grid
fig, axes = plt.subplots(1, 2, figsize=(10, 3))
# grid padrão
axes[0].plot(x1, x2, lw = 2)
axes[0].grid(True)

# Grid customizado
axes[1].plot(x1, x2, lw = 3)
axes[1].grid(color = 'b', alpha = 0.5, linestyle = 'dashed', linewidth = 0.5)
```

```python
# gráfico de linhas gêmeas
# 
fig, ax1 = plt.subplots()

ax1.plot(x1, x1, lw = 2, color = 'blue')
ax1.set_ylabel('Area', fontsize=18, color='blue')
for label in ax1.get_yticklabels():
    label.set_color('blue')
    
ax2 = ax1.twinx()
ax2.plot(x2, x2, lw = 2, color = 'red')
ax2.set_ylabel('Volume', fontsize=18, color='red')
for label in ax2.get_yticklabels():
    label.set_color('red')
```

```python
from numpy import random
# difetentes estilos de plots
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
# gráfico de dispensar
axes[0].scatter(xx, xx + 0.25 * random.randn(len(xx)))
# set titulo
axes[0].set_title('Scatter')

# grid
axes[1].step(n, n**2, lw=2)
axes[1].set_title('step')

# barras
axes[2].bar(n, n**2, align='center', width=0.5, alpha=0.5)
axes[2].set_title('bar')

# fill_between - prencha entre os limites
# não está suportando operações com exponênciação

axes[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
axes[3].set_title('fill_beetwen')
```

```python
# histogramas
n = np.random.randn(1000000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(n)
axes[0].set_title('Histograma Padrão.')
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title('Histograma comulativo.')
axes[1].set_xlim((min(n), max(n)));
```

```python
# color map
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def color_map(phi_m, phi_p):
    return( + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2 * phi_p))

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
x,y = np.meshgrid(phi_p, phi_m)
z = color_map(x, y).T

fig, ax = plt.subplots()

p = ax.pcolor(x / (2 * np.pi), y / (2 * np.pi), z, vmin=abs(z).min(), vmax=abs(z).max())
cb = fig.colorbar(p, ax=ax)
```
