```python
# importando pacote numpy
# importando pacote pandas
# importando random 
# importando matplotlib módulo pyplot
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# !pip install seaborn
# importando pacote seaborn
# seaborn roda sobre o matplotlib

import seaborn as sea
```

```python
# carregando um dos Datasets que vem com o Seaborn
dados = sea.load_dataset('tips')
```

```python
# total da conta em um restaurante
# tip é a gorjeta
# size quantidade de pessoas total na mesa

dados.head()
```

```python
# o método joinplot cria plot de 2 variável com gráficos bivariados
# e univariados
# bill - conta 
# tip - gorjeta
sea.jointplot('total_bill', 'tip', dados, kind = 'reg')
```

### O joinplot tem vários gráficos em 1 só
2 - histogramas 1 em cada eixo
1 - scatterplot
Mostrando na linha azul a linha de regressão
Ás duas áreas sombreadas de azul é chamado de zona de confiança - margem de erro.
Mostra exatamente as previsões de acordo com o total de dados.

```python
# método lmplot() cria plot com dados e modelos de regressão
sea.lmplot('total_bill', 'tip', dados, col='smoker')
```

Gráficos de reversão com variáveis diferentes
smoker = yes - fumante
smoker = no - não fumantes
Você conseguer comparar gráficos com tipos de dados rápidamente

```python
# construindo um dataframe
df = pd.DataFrame()
```

```python
# alimentando o DataFrame
df['a'] = random.sample(range(1, 100), 25)
df['b'] = random.sample(range(1, 100), 25)
```

```python
df.head()
```

```python
# Scatter Plot
sea.lmplot('a', 'b', data = df, fit_reg = True)
```

```python
# Density Plot
sea.kdeplot(df.b)
```

```python
# Density Plot
sea.kdeplot(df.b, df.a)
```

```python
# dist Plot
sea.distplot(df.a)
```

```python
# Histograma
plt.hist(df.a, alpha = .3)

# traçando a parte inferior
sea.rugplot(df.a)
```

```python
# Box Plot
sea.boxplot([df.b, df.a])
```

```python
# violin Plot
# Plot no formato de violino

sea.violinplot([df.a, df.b])
```

```python
# HeatMap
# mapa de calor mostra as desindades dos dados
# de acordo com a coloração

sea.heatmap([df.a, df.b], annot=True, fmt = 'd')
```

```python
# Clustermap
# Um grupo de dados similares, pegar um grande conjunto de dados
# criar clusters, pequenos agrupamentos depois coloco em um clustermap
# para ver como estão distribuidos

sea.clustermap(df)
```

## Outros Pots

```python
sea.set(style='darkgrid')
dados = sea.load_dataset('tips')
q = sea.FacetGrid(dados, row='sex', col='time', margin_titles=True)
bins = np.linspace(0, 60, 13)
q.map(plt.hist, 'total_bill', color='steelblue', bins = bins, lw = 0)
```

Temos em baixo o total_bill<br>
time = lunch - almoço<br>
time = dinner - jantar<br>
Nas laterais<br>
sex - Male - masculino<br>
sex - Female - feminino<br>
Isto nos permite comparar 4 tipo de variáveis.
Se apresentar isto aqui para pessoas sem conhecimento tecnico pode complicar ainda mais.

```python
# diversos plots simultaneos
sea.set(style='white', palette='muted')
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sea.despine(left=True)

rs = np.random.RandomState(10)

b, g, r, p = sea.color_palette('muted', 4)

d = rs.normal(size=100)

sea.distplot(d, kde=False, color=b, ax = axes[0, 0])
sea.distplot(d, hist=False, rug=True, color=r, ax = axes[0, 1])
sea.distplot(d, hist=False, color=g, kde_kws={'shade': True}, ax = axes[1, 0])
sea.distplot(d, color=p, ax=axes[1, 1])

plt.setp(axes, yticks = [])
plt.tight_layout()
```

```python
# plots com distribuições marginais
from scipy.stats import kendalltau
sea.set(style='ticks')
rs = np.random.RandomState(11)
x = rs.gamma(2, size = 1000)
y = -.5 * x + rs.normal(size = 1000)
sea.jointplot(x, y, kind='hex', stat_func=kendalltau, color='#4CB391')
```

### Gráfico mostra uma relação negativa entre os dados
Os dados representa uma contra barra que começa no canto superior esquerdo e termina no canto superior direito

```python
# regressão logística
sea.set(style='darkgrid')
df = sea.load_dataset('titanic')
pal = dict(male='#6495ED', female='#F08080')
g = sea.lmplot('age', 'survived', col='sex', hue='sex', data=df, palette=pal, y_jitter=.02, logistic=True)
g.set(xlim=(0, 80), ylim=(-.05, 1.05));
```

```python
# Pair plot
sea.set(style='darkgrid')
df = sea.load_dataset('iris')
sea.pairplot(df, hue='species', size=2.5)
```
