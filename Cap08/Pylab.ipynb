## Paylab<br><h6>Ele é uma junção entre pyplot do matplotlib<br>E NumPy<br>Nôs da uma caracteristicas interessante para gráficos.</h6>

```python
# visualizando gráficos dentro do Jupyter Notebook
# O pylab combina funcionalidades do pyplot com funcionalidades do Numpy
from pylab import *
%matplotlib inline
```

```python
# criando o gráfico
# linha espace start, stop, steps, do pacote NumPy apartir do pacote pylab
# definindo eixos

x = linspace(0, 5, 10)
y = x ** 2

# definindo objeto 
fig = plt.figure()
# definindo eixos
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# modelando gráfico
axes.plot(x, y, 'r')

# definindo eixos
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha');
```

```python
# gráfico com 2 figuras
# figura secundaria dentro da figura principal

x = linspace(0, 5, 10)
y = x ** 2

# criando um objeto fig
fig = plt.figure()

# definindo eixos
# eixo da figura principal
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# eixo da figura secundária
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# figura secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundária')
```

```python
# gráfico em paralelo
# este é util se quiser compara a mesma variavel y com x

fig, axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Titulo')
    
fig.tight_layout()
```
