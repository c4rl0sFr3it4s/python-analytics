## Gráficos 3D

Interpretar um gráfico 3D está em um nível maior de compreendimento de muitos usuários.
Precisa realmente ter um conhecimento dos dados melhor, porém quando 
se passa para um usuário um gráfico 3D, é como uma obrigação ele conhecer aquilo com mais detalhes. 

```python
# ferramenta do pacote matplotlib
# módulo mpl_toolkits.mplot3d.axes3d
# função Axes3D
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
%matplotlib inline
```

```python
# color map
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def color_map(phi_m, phi_p):
    return( + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2 * phi_p))

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
x,y = np.meshgrid(phi_p, phi_m)
z = color_map(x, y).T

fig, ax = plt.subplots()

p = ax.pcolor(x / (2 * np.pi), y / (2 * np.pi), z, vmin=abs(z).min(), vmax=abs(z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
# Gráfico 3D surface
fig = plt.figure(figsize=(14, 6))

# projeção 3d
ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.plot_surface(x, y, z, rstride=4, cstride=4, linewidth=0)

ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(x, y, z, rstride=1, cstride=1, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

```python
# wire frame
# definindo o tamanho do desenho
# bastante comum em ciencia de dados

fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_wireframe(x, y, z, rstride=4, cstride=4)
```

```python
# coutour plot com projeção
fig = plt.figure(figsize=(8, 6))
pi = 3.14
ax = fig.add_subplot(1, 1, 1, projection='3d')

ax.plot_surface(x, y, z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(x, y, z, zdir='z', offset=pi)
cset = ax.contour(x, y, z, zdir='x', offset=pi)
cset = ax.contour(x, y, z, zdir='y', offset=3*pi)

ax.set_xlim3d(-pi, 2 * pi);
ax.set_ylim3d(0, 3 * pi);
ax.set_zlim3d(-pi, 2*pi);
```
