## Gráficos

```python
# importando pacotes, os, sqlite3, matplotlib, datetime, time
import os
import sqlite3
import datetime
import time
import matplotlib.pyplot as plt
# plotar gráfico inline
%matplotlib notebook
```

```python
# removendo arquivo db(caso exista)
# não sendo necessário para uso do mesmo arquivo
os.remove('informatica.db') if os.path.exists('informatica.db') else None
```

```python
# criando conexão
conn = sqlite3.connect('informatica.db')

# criando cursor
c = conn.cursor()
```

```python
# funções
def close():
    '''
        fechar conexão, cursor
    '''
    c.close()
    conn.close()
    
def create_table():
    '''
        criar tabela e atributos
    '''
    c.execute('CREATE TABLE IF NOT EXISTS produtos'\
              '(id INTEGER PRIMARY KEY AUTOINCREMET NOT NULL,'\
              'date TEXT,'\
              'prod_nome TEXT, '\
              'valor REAL)')

def data_insert(prod, value):
    '''
        inserir domínios
    '''
    new_date = datetime.datetime.now()
    new_prod = prod
    new_value = value
    c.execute("INSERT INTO produtos(date, prod_nome, valor)"\
            "VALUES(?, ?, ?)", (new_date, new_prod, new_value))
    conn.commit()
    
def data_select_all():
    '''
        selecionar e mostrar domínios
    '''
    c.execute('SELECT * FROM produtos')
    print('Produtos')
    for line in c.fetchall():
            print('Id: %s, Data: %s, Nome: %s, Valor: %s' %line)
    print('-----------------------------------------')

def data_update_prod():
    '''
        atualiza domínio produto
    '''
    id = int(input('Número ID: '))
    produto = input('Nome do novo produto: ')
    c.execute('UPDATE produtos SET prod_nome ='+ produto+ ' WHERE id=?', (id))
    
def data_update_value():
    '''
        atualizar domínio value
    '''
    id = int(input('Número ID: '))
    valor = float(input('Valor R$: '))
    c.execute('UPDATE produtos SET valor ='+ valor+ 'WHERE id=?', (id))
    conn.commit()
    
def data_remove():
    '''
        remover domínio
    '''
    id = int(input('Número do ID: '))
    c.execute('DELETE FROM produtos WHERE id=?', (id))
    conn.commit()
    
def data_select_specific(pos):
    '''
        selecionar atributo especifico
    '''
    c.execute('SELECT * FROM produtos')
    for line in c.fetchall():
        print(line[pos])
    
def data_grafic():
    c.execute('SELECT * FROM produtos')
    ids = []
    valores = []
    data = c.fetchall()
    for line in data:
        ids.append(line[0])
        valores.append(line[3])
        
    plt.bar(ids, valores)
    plt.show()
```

```python
# mostrando dados
data_select_all()
```

```python
# gerando gráficos
data_grafic()
```
